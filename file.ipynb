```python
# Ensure the file path is correct
df_users = pd.read_csv(r'C:\Users\KIIT0001\Downloads\IMDb Movies India.csv', sep="::", engine="python", encoding='ISO-8859-1')

# Drop rows with missing values
df_users.dropna(inplace=True)

# Display the first 10 rows
df_users.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

# Ensure column names are clean
df_users.columns = df_users.columns.str.strip()

# Check if the 'Gender' column exists
print(df_users.columns)

# Proceed with label encoding if 'Gender' column exists
if 'Gender' in df_users.columns:
    labelencoder = LabelEncoder()
    df_users['Gender'] = labelencoder.fit_transform(df_users['Gender'])
    print(df_users.head())
else:
    print("The 'Gender' column is not found.")
```

```python
df_users.isna().sum()
```

```python
df_data=pd.concat([df_movie,df_ratings,df_users],axis=1)
df_data.dropna()
df_data.head(10)
```

```python
df_data.shape
```

```python
# Ensure column names are clean
df_data.columns = df_data.columns.str.strip()

# Check the columns to verify if they exist
print(df_data.columns)

# Drop columns if they exist, ignore missing ones
df2 = df_data.drop(["Occupation", "Zip-code", "Timestamp"], axis=1, errors='ignore')

# Display the first few rows of the updated DataFrame
print(df2.head())
```

```python
df_final=df2.dropna()
```

```python
df_final.shape
```

```python
# Check the columns of df_final
print(df_final.columns)

# Ensure 'Gender' column is present and strip any unwanted spaces
df_final.columns = df_final.columns.str.strip()

# If 'Gender' exists, proceed with the countplot
if 'Gender' in df_final.columns:
    sns.countplot(x='Gender', hue='Ratings', data=df_final)
    plt.show()
else:
    print("The 'Gender' column is missing or misnamed.")
```

```python
# Check the columns of df_final
print(df_final.columns)

# Ensure 'Age' column is present and strip any unwanted spaces
df_final.columns = df_final.columns.str.strip()

# If 'Age' exists, plot the histogram
if 'Age' in df_final.columns:
    df_final['Age'].plot.hist(bins=25)
    plt.ylabel("Frequency")  # Rename y-axis to "Frequency"
    plt.xlabel("Age")
    plt.title("Distribution of Age")
    plt.show()
else:
    print("The 'Age' column is missing or misnamed.")
```

```python
# Check the columns of df_final
print(df_final.columns)

# Remove leading/trailing spaces from column names
df_final.columns = df_final.columns.str.strip()

# Verify if the 'Ratings' column exists, and then plot
if 'Ratings' in df_final.columns:
    df_final['Ratings'].value_counts().plot(kind='bar')
    plt.xlabel("Ratings")
    plt.ylabel("Count")
    plt.title("Distribution of Ratings")
    plt.show()
else:
    print("The 'Ratings' column is missing or misnamed.")
```
